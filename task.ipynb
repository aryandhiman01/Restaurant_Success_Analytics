```python
# Outlier Detection Using IQR

Q1 = df_clean[
    "Aggregate rating"
].quantile(0.25)

Q3 = df_clean[
    "Aggregate rating"
].quantile(0.75)

IQR = Q3 - Q1

lower_bound = (
    Q1 - 1.5 * IQR
)

upper_bound = (
    Q3 + 1.5 * IQR
)

outliers = df_clean[
    (
        df_clean[
            "Aggregate rating"
        ] < lower_bound
    )
    |
    (
        df_clean[
            "Aggregate rating"
        ] > upper_bound
    )
]

print(
    "Total Outliers:",
    len(outliers)
)
```

```python
# Boxplot

plt.figure(figsize=(10,4))

sns.boxplot(
    x=df_clean[
        "Aggregate rating"
    ]
)

plt.title(
    "Rating Outlier Detection"
)

plt.show()
```

```python
# Phase 10 : Feature Engineering
# Objective

# Feature Engineering is the process of creating new meaningful variables from existing data to improve business understanding and analysis.
```

```python
# Cost Category

def cost_category(cost):

    if cost <= 500:
        return "Budget"

    elif cost <= 1500:
        return "Moderate"

    elif cost <= 5000:
        return "Premium"

    else:
        return "Luxury"


df_clean["Cost Category"] = df_clean[
    "Average Cost for two"
].apply(cost_category)
```

```python
df_clean[
    "Cost Category"
].value_counts()
```

```python
# Rating Category

def rating_category(rating):

    if rating < 2:
        return "Poor"

    elif rating < 3.5:
        return "Average"

    elif rating < 4.5:
        return "Good"

    else:
        return "Excellent"


df_clean["Rating Category"] = df_clean[
    "Aggregate rating"
].apply(rating_category)
```

```python
df_clean[
    "Rating Category"
].value_counts()
```

```python
# Cost Category vs Rating

cost_rating = (

    df_clean
    .groupby(
        "Cost Category"
    )
    [
        "Aggregate rating"
    ]
    .mean()

)

cost_rating
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=df_clean,
    x="Cost Category",
    y="Aggregate rating",
    order=[
        "Budget",
        "Moderate",
        "Premium",
        "Luxury"
    ]
)

plt.title(
    "Average Rating by Cost Category"
)

plt.show()
```

```python
# Cost Category vs Votes

cost_votes = (

    df_clean
    .groupby(
        "Cost Category"
    )
    [
        "Votes"
    ]
    .mean()

)

cost_votes
```
