```python
# # Phase 3: Descriptive Statistics

# ## Objective

# Descriptive statistics provide a summary of the dataset's numerical variables.

# These measures help understand:

# - Central Tendency
# - Data Distribution
# - Data Variability
# - Presence of Outliers
```

```python
numerical_cols = [
    "Average Cost for two",
    "Aggregate rating",
    "Votes",
    "Price range"
]

for col in numerical_cols:

    print("\n" + "="*50)
    print("Column:", col)

    print("Mean:", round(df_clean[col].mean(),2))
    print("Median:", round(df_clean[col].median(),2))
    print("Mode:", df_clean[col].mode()[0])
    print("Standard Deviation:", round(df_clean[col].std(),2))
    print("Variance:", round(df_clean[col].var(),2))
```

```python
# Create Statistics Table
stats_table = pd.DataFrame({
    "Mean": df_clean[numerical_cols].mean(),
    "Median": df_clean[numerical_cols].median(),
    "Mode": [
        df_clean[col].mode()[0]
        for col in numerical_cols
    ],
    "Std Dev": df_clean[numerical_cols].std(),
    "Variance": df_clean[numerical_cols].var()
})

stats_table
```

```python
# 1. Average Cost for Two : The average restaurant cost for two people is approximately ₹1199, while the median cost is only ₹400. The large difference between mean and median indicates the presence of extremely expensive restaurants.
# 2. Aggregate Rating : The average restaurant rating is approximately 2.67, while the median rating is 3.2. Most restaurants fall within the average to good rating range.
# 3. Votes : Restaurants receive an average of 157 votes, while the median is only 31 votes.
# 4. Price Range : The average price range is approximately 1.8, indicating that most restaurants belong to lower or moderate pricing categories.
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df_clean[
        [
            "Average Cost for two",
            "Aggregate rating",
            "Votes",
            "Price range"
        ]
    ]
)

plt.title("Distribution of Numerical Features")
plt.xticks(rotation=20)

plt.show()
```

```python
# ## Phase 3 Conclusion

# Descriptive statistical analysis revealed significant variation in restaurant costs and customer votes.

# Restaurant ratings showed moderate variation, while price ranges remained relatively concentrated.

# The analysis also indicated the presence of potential outliers, particularly in restaurant costs and customer engagement metrics, which will be explored further in subsequent phases.
```

```python
# # Phase 5: Cuisine Analysis

# ## Objective

# This phase focuses on identifying the most popular cuisines available in the dataset and understanding customer food preferences.A
```

```python
# Extract Individual Cuisines

from collections import Counter

all_cuisines = []

for cuisine in df_clean["Cuisines"].dropna():

    cuisine_list = cuisine.split(",")

    all_cuisines.extend(
        [c.strip() for c in cuisine_list]
    )
```

```python
# Top 10 Most Popular Cuisines

top_10_cuisines_df = pd.DataFrame(
    top_10_cuisines,
    columns=[
        "Cuisine",
        "Count"
    ]
)

top_10_cuisines_df
```

```python
# Visualization
plt.figure(figsize=(12,6))

sns.barplot(
    data=top_10_cuisines_df,
    x="Count",
    y="Cuisine"
)

plt.title(
    "Top 10 Most Popular Cuisines"
)

plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")

plt.show()
```

```python
# Percentage Contribution

top_10_cuisines_df[
    "Percentage"
] = round(
    (
        top_10_cuisines_df["Count"]
        /
        len(all_cuisines)
    ) * 100,
    2
)

top_10_cuisines_df
```

```python
# Observation

# The cuisine distribution analysis highlights the most frequently served cuisines across restaurants.

# Popular cuisines generally reflect customer demand and cultural food preferences.
```

```python
# Highest Rated Cuisines

cuisine_rating = (
    df_clean
    .groupby("Cuisines")
    ["Aggregate rating"]
    .mean()
    .sort_values(
        ascending=False
    )
    .head(10)
)

cuisine_rating
```

```python
cuisine_rating_df = pd.DataFrame(
    cuisine_rating
)

cuisine_rating_df
```

```python
# Highest Rated Cuisine Visualization
plt.figure(figsize=(12,6))

sns.barplot(
    x=cuisine_rating_df[
        "Aggregate rating"
    ],
    y=cuisine_rating_df.index
)

plt.title(
    "Top Rated Cuisine Categories"
)

plt.xlabel("Average Rating")
plt.ylabel("Cuisine")

plt.show()
```

```python
# Observation

# Certain cuisine categories consistently receive higher customer ratings.

# This indicates strong customer satisfaction and positive dining experiences associated with these cuisines.
```
